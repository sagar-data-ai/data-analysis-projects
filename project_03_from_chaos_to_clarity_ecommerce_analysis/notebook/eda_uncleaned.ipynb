**BIVARIATE ANALYSIS**

```python
# 1  Analysis between price and freight_value
sns.scatterplot(data=df, x='price', y='freight_value')
plt.title('Price vs Freight Value')
plt.show()


# 2  Analysis between product_category_name_english and review_score
sns.boxplot(data=df, x='product_category_name_english', y='review_score')
plt.xticks(rotation=90)
plt.title('Product Category vs Review Score')
plt.show()


# 3  Analysis between payment_type and review_score
sns.boxplot(data=df, x='payment_type', y='review_score')
plt.title('Payment Type vs Review Score')
plt.show()


# 4  Analysis between customer_state and freight_value
sns.barplot(data=df, x='customer_state', y='freight_value', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Freight Value by Customer State')
plt.show()


# 5  Analysis between order_item_id and payment_value
sns.scatterplot(data=df, x='order_item_id', y='payment_value')
plt.title('order item id vs Payment Value')
plt.show()
```

```python
# 6  Analysis between price and review_score
sns.boxplot(data=df, x='review_score', y='price')
plt.title('Review Score vs Product Price')
plt.show()


# 7  Analysis between product_category_name_english and payment_value
sns.barplot(data=df, x='product_category_name_english', y='payment_value', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Payment Value by Product Category')
plt.show()


# 8  Analysis between seller_state and freight_value
sns.barplot(data=df, x='seller_state', y='freight_value', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Freight Value by Seller State')
plt.show()


# 9  Analysis between payment_type and payment_value
sns.boxplot(data=df, x='payment_type', y='payment_value')
plt.title('Payment Type vs Payment Value')
plt.show()


# 10  Analysis between review_score and freight_value
sns.boxplot(data=df, x='review_score', y='freight_value')
plt.title('Freight Value vs Review Score')
plt.show()
```

```python
# 11  Analysis between customer_state and payment_value
sns.barplot(data=df, x='customer_state', y='payment_value', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Payment Value by Customer State')
plt.show()


# 12  Analysis between product_weight_g and freight_value
sns.scatterplot(data=df, x='product_weight_g', y='freight_value')
plt.title('Product Weight vs Freight Value')
plt.show()


# 13  Analysis between product_category_name_english and freight_value
sns.barplot(data=df, x='product_category_name_english', y='freight_value', estimator='mean')
plt.xticks(rotation=90)
plt.title('Average Freight Value by Product Category')
plt.show()


# 14  Analysis between payment_type and price
sns.boxplot(data=df, x='payment_type', y='price')
plt.title('Payment Type vs Product Price')
plt.show()


# 15  Analysis between seller_state and review_score
sns.boxplot(data=df, x='seller_state', y='review_score')
plt.xticks(rotation=90)
plt.title('Seller State vs Review Score')
plt.show()
```

```python
# 16  Analysis between order_status and review_score
sns.boxplot(data=df, x='order_status', y='review_score')
plt.title('Order Status vs Review Score')
plt.show()


# 17  Analysis between price and payment_value
sns.scatterplot(data=df, x='price', y='payment_value')
plt.title('Price vs Payment Value')
plt.show()


# 18  Analysis between freight_value and payment_value
sns.scatterplot(data=df, x='freight_value', y='payment_value')
plt.title('Freight Value vs Payment Value')
plt.show()


# 19  Analysis between product_category_name_english and order_status
sns.countplot(data=df, x='product_category_name_english', hue='order_status')
plt.xticks(rotation=90)
plt.title('Order Status Distribution by Product Category')
plt.show()


# 20  Analysis between product_weight_g and price
sns.scatterplot(data=df, x='product_weight_g', y='price')
plt.title('Product Weight vs Product Price')
plt.show()
```
