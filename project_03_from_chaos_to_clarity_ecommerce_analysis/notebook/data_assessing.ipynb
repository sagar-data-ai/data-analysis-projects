###  Handling Multiple Rows for a Single Order

- All rows belonging to the same `order_id` were grouped together.  
- For each order:
  - The `payment_value` values were **summed** to get the total amount paid.
  - The `payment_sequential` was replaced with its **maximum value** to represent the total number of transactions made.
  - The `payment_type` was marked as the actual payment type if all transactions used the same method, otherwise labeled as **"multiple"**.
- As a result, every order now exists as a **single row**, representing complete payment information in one record instead of having multiple rows.


### üì¶ Transforming `order_item_id` into `items_in_order`
After converting multple rows of one order into **one single row**   
keeping multiple `order_item_id` values was no longer meaningful.

- All rows corresponding to the same `order_id` were grouped together.
- The maximum value of `order_item_id` will be stored in a new column **`items_in_order`** .

**Outcome:**  
The dataset now includes a single intuitive feature ‚Äî `items_in_order` ‚Äî indicating the total number of products per order.
